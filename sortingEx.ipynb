```python
def freq_selection_sort(freq_dict, nums):
    n = len(nums)
    counter = 0
    print('iteration:', counter, 'List:', nums)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            # Compare based on frequency first, then by value (decreasing order)
            if (freq_dict[nums[j]], -nums[j]) < (freq_dict[nums[min_idx]], -nums[min_idx]):
                min_idx = j
            counter = counter+1
        # Swap the found minimum element with the current element
        nums[i], nums[min_idx] = nums[min_idx], nums[i]
        
        print('iteration:', counter, 'List:', nums)
    print('Using selection Sort: complete at', counter+1)

    return counter, nums

def freq_insertion_sort(freq_dict, nums):
    counter = 0
    n = len(nums)
    print('iteration:', counter, 'List:', nums)
    for i in range(1, n):
        key = nums[i]
        j = i - 1

        # Compare based on frequency, then value in decreasing order
        while j >= 0 and (freq_dict[nums[j]], -nums[j]) > (freq_dict[key], -key):
            nums[j + 1] = nums[j]
            j -= 1

        nums[j + 1] = key
        counter = counter + 1
        print('iteration:', counter, 'List:', nums)
    print('Using insertion Sort: complete at', counter+1)
    return counter, nums



def frequencySort(nums):

    freq_dict = {}
    sorted_list = []
    if nums is None:
        return False
    for num in nums:
        if num in freq_dict:
            freq_dict[num] = freq_dict[num] +1
        else:
            freq_dict[num] = 1

    freq_selection_sort(freq_dict, nums)
    print('Start Insertion sort')
    freq_insertion_sort(freq_dict,nums)
        
    

```
